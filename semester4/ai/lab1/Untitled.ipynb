```python
import random
def spinner(probabilityDistribution):
    sum_distribution = 0
    random_number = random.random()
    i = 0
    for probability in probabilityDistribution:
        sum_distribution += probability
        if random_number < sum_distribution:
            return i
        else:
            i = i+1
probabilityDistribution = [0.1, 0.2, 0.3, 0.4]
print(spinner(probabilityDistribution))
```

```python
import random
def spinner(probabilityDistribution):
    n = len(probabilityDistribution)
    for i in range(n):
        probabilityDistribution[i]=round(probabilityDistribution[i]/360, 4);
    sum_distribution = 0
    random_number = random.random()
    i = 0
    for probability in probabilityDistribution:
        sum_distribution += probability
        if random_number < sum_distribution:
            return i
        else:
            i = i+1
p = [45,85,70,160]
print(spinner(p))
```

```python
import random
import math

radius_pipe = 30  
len_square_yard = 100
radius_ball = 1

def falls_inside(x, y):
    distance = math.sqrt((x - len_square_yard/2)**2 + (y - len_square_yard/2)**2)
    return distance <= radius_pipe

total_balls = 1000000
inside_pipe_count = 0

for _ in range(total_balls):
    x = random.uniform(0, len_square_yard)
    y = random.uniform(0, len_square_yard)
    if falls_inside(x, y):
        inside_pipe_count += 1

experimental_probability = inside_pipe_count / total_balls

mathematical_probability = math.pi * radius_ball**2 / len_square_yard**2

print("the experimental probability is -->", experimental_probability)
print("the mathematical probability is -->", mathematical_probability)
```
